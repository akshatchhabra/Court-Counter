process_or_store() implementation:
this method opens the JSONline file and appends each tweet to the file in JSON format

```
def process_or_store(tweet):
  with jsonlines.open('tweets.jsonl', mode='a') as file:
    file.write(tweet)
```

Opening JSONLine file for reading and obtaining the required fields

```
alltweets = []
with open('tweets.jsonl') as file:
    for line in file:
        j_content = json.loads(line)
        l = []
        l.append(j_content["text"])
        l.append(j_content["created_at"][:19])
        l.append(j_content["favorite_count"])
        l.append(j_content["retweet_count"])
        try:
          count = 0
          if "extended_entities" in j_content:
            count += len(j_content["extended_entities"]["media"])
          if "entities" in j_content:
            count += len(j_content["entities"]["media"])
          l.append(count)
        except:
          l.append("None")
        alltweets.append(l)
```

Creating a table for displaying tweets

```
from prettytable import PrettyTable

x = PrettyTable(hrules = 1)

x.field_names = [ "Text", "Date and Time", "Favorites/Likes", "Retweets", 
                 "Number of Images"]

for l in alltweets:
  x.add_row(l)
print(x)
```
